**Цель исследования**

Изучить рынок российского кинопроката и выявить текущие тренды, кроме того необходимо узнать, влияет ли как-то государственная поддержка на рейтинги фильма

**Данные**

Набор данных содержит информацию о прокатных удостоверениях, сборах и государственной поддержке фильмов, а также информацию с сайта КиноПоиск.

-mkrf_shows.csv — данные о прокате в российских кинотеатрах.
-mkrf_movies.csv — данные о прокатных удостоверениях.

**Описание переменных в данных**
Таблица shows
- `box_office` - кассовые сборы
- `puNumber` - номер прокатного удостоверения*

Таблица movies
- `puNumber` - номер прокатного удостоверения*
- `title` - Название фильма. Так фильм идентифицируется и продвигается среди публики.
- `show_start_date` - Дата первого показа фильма в кинотеатрах или его выпуска для публики. Это может указывать его актуальность на рынке.
- `type` - Категория фильма: художественный, научно-популярный, документальный и т.д. Дает представление о содержании фильма и целевой аудитории.,
- `film_studio` - Название студии, которая сняла фильм. Это может указывать на качество фильма и его бюджет.
- `production_country` - Страна производства фильма. Может влиять на его стиль, язык и культурный контекст.
- `director` - Режиссер фильма, человек чье видение и стиль работы могут значительно повлиять на конечный продукт.
- `producer` - Продюсер фильма, человек отвечающий за финансовую и административную стороны производства.
- `age_restriction` - Возрастное ограничение фильма, может влиять на его аудиторию и тему.
- `refundable_support` - Объём возвратных средств государственной поддержки.
- `nonrefundable_support` - Объём невозвратных средств государственной поддержки.
- `budget` - Общий бюджет фильма, включая все расходы на производство.
- `financing_source` - Источник государственного финансирования.
- `ratings` - Рейтинг фильма на КиноПоиске.
- `genres` - Жанры фильма.

*Прока́тное удостовере́ние фи́льма — документ, дающий право на прокат (демонстрацию) фильма на территории Российской Федерации, а также на его тиражирование и распространение, в том числе по кабельным или эфирным каналам.

#### <a id='toc1_2_1_1_'></a>[Создание функции контролируюшей повторный запуск ячейки](#toc0_)

Если код нельзя запускать второй раз, сюда будет добавлена переменная, которая будет контролировать повторный запуск

```python
# Хранилище переменных, контролирующих повторный запуск кода
cell_run_flags = set()


# Функция для запуска кода один раз
def run_code_only_once(code, flag_name, force_run=False):
    global cell_run_flags
    if force_run:
        code()
        return True
    elif not flag_name in cell_run_flags:
        code()
        return True
    else:
        message = f"Cообщение. Основной код ячейки уже был запущен при предыдущем запуске, перезагрузите тетрадь для повторного запуска ячейки или вызовите функцию  run_code_only_once со значением force_run=True. При повторном запуске ячейки без перезагрузки тетради могут быть ошибки"
        print(message)
        return False
```

Напишем функцию, которая выводит строки с одинаковыми значениями в столбце

```python
def return_df_by_column_same_values(data, characteristic):
    # Создадим массив. В нем будут повторяющиеся значения столбца
    same_characteristics_list = data[data[characteristic].duplicated()][
        characteristic
    ].tolist()
    # Выберем строки с повторяющимися значениями столбца
    same_characteristics_rows = data[
        data[characteristic].isin(same_characteristics_list)
    ]

    return same_characteristics_rows


return_df_by_column_same_values(data, "title").sort_values(by="title").head(5)
```

Видим, что функция работает корректно, также дополнительно проверил, что количество строк в таблице data с повторяющимся значением совпадает с количеством строк в таблице `same_characteristics_rows` с таким же значением. Проверку функции для читаемости кода опустил.

#### <a id='toc1_3_2_1_'></a>[Смена типа данных в столбцах `refundable_support`, `nonrefundable_support`, `budget`, `box_office`](#toc0_)

Так как в столбцах связанных с деньгами: `refundable_support`, `nonrefundable_support`, `budget`, `box_office` тип данных стоит сменить на целое число так как дробная часть слишком мала, чтобы иметь вес при таких больших числах, сменим тип данных в этих столбцах на целочисленный.

```python
def convertable_to_numeric(value):
    try:
        pd.to_numeric(value)
        return True
    except:
        return False


# Функция для смены типа данных в числовой тип. Если в столбце нет NaN значений и ошибок, возвращает два датафрейма: с измененным типом данных и series размером с dataframe где нет ни одного значения False.
# Если в столбце есть NaN значения или значения, которые нельзя преобразовать в целочисленный тип, возвращает
# 1) dataframe со всеми возможно преобразоваными значениями в столбце, оставляя NaN и значения, которые нельзя преобразовать в целочисленный тип
# 2) series, в котором напротив строки с не NaN значения, которое не вышло преобразовать стоит False
# При type = 'int' преобразует колонку в целочисленый тип данных, при других значениях возвращает датафрейм где обрабатываемая колонка принимает числовой тип данных, по умолчанию float
# Функционал присвоения типа данных int проверен на примере и работает верно


def convert_column_to_number_safely(
    data, *columns, cant_be_changed_show_number=10, type="float"
):
    for column in columns:
        # Список не NaN значений, которые можно преобразовать в целочисленный тип
        can_be_changed = data.loc[:, column].apply(convertable_to_numeric)

        # Список не NaN значений, которые нельзя преобразовать в целочисленный тип
        cant_be_changed = data[column].notna() & ~can_be_changed

        # Смена типа данных на числовой в строках без NaN
        data.loc[can_be_changed, column] = pd.to_numeric(
            data.loc[can_be_changed, column], errors="raise"
        )

        # Смена типа данных на целочисленный в строках без NaN, если type = 'int'
        if type == "int":
            data.loc[can_be_changed & data[column].notna(), column] = data.loc[
                can_be_changed & data[column].notna(), column
            ].astype("int64")

        # Вывод строк количества строк содержащих NaN значения,
        print(
            f"Количество строк в столбце {column} содержащих NaN значения = {data[column].isna().sum()}. Если NaN значения присутствуют и передан аргумент int, тип данных все равно будет float \n"
        )

        # Вывод строк, которые нельзя преобразовать в целочисленный тип
        if cant_be_changed.sum() > 0:
            print(
                f"Эти значения нельзя преобразовать в целочисленный тип, их количество = {cant_be_changed.sum()}, показано {cant_be_changed_show_number} из них"
            )
            print(data.loc[cant_be_changed, column].head(cant_be_changed_show_number))

        # Смена типа данных всего столбца
        if cant_be_changed.sum() == 0:
            if type == "int" and data[column].isna().sum() == 0:
                data[column] = data[column].astype("int64")
            else:
                data[column] = pd.to_numeric(data[column], errors="coerce")

    return data, cant_be_changed


# Смена типа данных в столбцах `refundable_support`, `nonrefundable_support`, `budget`, `box_office`
data, cant_be_changed = convert_column_to_number_safely(
    data,
    "refundable_support",
    "nonrefundable_support",
    "budget",
    "box_office",
    cant_be_changed_show_number=10,
    type="int",
)
```

**Справочная информация**

Редкие и аномальные значения мы отбрасываем для того, чтобы данные были более качественными для исследования самых частых случаев. ### <a id='toc1_3_5_'></a>[Преобразование данных](#toc0_)

**Справочная информация**

**Что такое one-hot кодирование и почему оно такое горячее?**

One-hot кодирование - это процесс преобразования категориальных столбцов в виде бинарных. Это делается для того, чтобы модели машинного обучения было проще обрабатывать категориальные данные. Вместо того, чтобы использовать одну колонку с категориальными данными, мы создаем столько колонок, сколько уникальных значений в категориальной колонке. Если значение в категориальной колонке равно значению в одной из колонок, то в этой колонке стоит 1, в остальных 0. Если в категориальной колонке несколько значений одновременно, то в нескольких колонках стоит 1, в остальных 0.

**Информация**
- В данном разделе будут предобработаны в том числе и столбцы в которых есть пропуски: `genres`, `producer`.

#### <a id='toc1_3_5_2_'></a>[Преобразование данных в столбце `film_studio`.](#toc0_)

Выведем первые 5 значений столбца

```python
data["film_studio"].head(5)
```

Видим, что каждому фильму соответсвует несколько киностудий, разделенных запятой. По этому столбцу невозможно сказать как участие какой-либо киностудии влияет на  фильма, ведь почти каждому фильму соответствует уникальный набор киностудий и, следовательно, уникальное значение столбца. Разделим эту характеристику на несколько столбцов в названии которой будет название киностудии, а в ячейке 1 если киностудия участвовала в создании фильма, 0 если не участвовала. Иначе говоря проведем One-hot кодирование.

Преобразование столбца `film_studio` в массив пригодный для One-hot кодирования

```python
# Массив для значений, которые не вышло преобразовать
film_studio_split_exceptions = []


# Функция убирающая вложенность массивов
def flatten_list(nested_list):
    flat_list = []
    for item in nested_list:
        if isinstance(item, list):
            flat_list.extend(flatten_list(item))
        else:
            flat_list.append(item)
    return flat_list


# Функция для разделения строк по знаку.
def split_by_sign(s, sign):
    if not isinstance(s, list):
        if sign in s:
            return s.split(sign)
        else:
            raise Exception("В строке нет знака или значение не строка")
    return s


# Функция для вызова другой функции и сохранения значений вызвающих ошибку в global_exception_list
def safe_process_value(s, global_exception_list, function, *args, **kwargs):
    try:
        return function(s, *args, **kwargs)
    except:
        global_exception_list.append(s)
        return s


# Применение функции разбиения по запятой
data["film_studio"] = data["film_studio"].apply(
    lambda x: safe_process_value(x, film_studio_split_exceptions, split_by_sign, ",")
)

# Вывод первых 10 необработанных значений
print(
    f"Первые 10 необработанных значений: {film_studio_split_exceptions[:10]}, их количество: {len(film_studio_split_exceptions)}. Если вы запускает ячейку второй раз, все необработанные значения были превращенны в массивы с 1 информацией"
)

# Вывод списка значений
data["film_studio"].head(10)
```

**Про господдержку**
Так как мы будем анализировать влияние господдержки, желательно указать, если, часто появляющаяся киностудия, получала господдержку или в съемках фильма участвовала другая киностудия, получавшая господдержку.

Создадим множество киностудий получивших господдержку, чтобы использовать его при проверке участия господдержанных киностудий в съемках фильмаю

```python
# Функция для запуска кода внутри через функцию, разрешающую запустить код только один раз за запуск тетради
def code_to_run():
    global gos_sup_studios_set
    global data

    # Преобразование столбца `film_studio` в массив
    gos_sup_film_studios_participate_list = data.loc[
        data["financing_source"] != "отсутствует", "film_studio"
    ].explode()
    gos_sup_studios = gos_sup_film_studios_participate_list.dropna().tolist()

    # Преобразование в множество для быстрой проверки вхождения
    gos_sup_studios_set = set(gos_sup_studios)

    # Вывод множества
    print(gos_sup_studios_set)


# Запуск кода code_to_run через функцию, разрешающую запустить код только один раз за запуск тетради. Сохранение результата в переменную, позволяющую сделать запись о том, что ячейка была запущена
gos_sup_studios_set_cell = run_code_only_once(
    code_to_run, "gos_sup_studios_set_cell", 1
)

# Добавления записи о том, что ячейка была запущена. Если бы в ходе выполнения были ошибки, последняя запись будет использована для вывода сообщения об ошибке
if gos_sup_studios_set_cell:
    cell_run_flags.add("gos_sup_studios_set_cell")
```

Теперь в каждой ячейке `film-studio` заменим редко встречающиеся киностудии на `другие`, а также добавим префикс `гп-` к часто встречающимся киностудиям, которые получали господдержку. Если в съемках фильма участвовала киностудия, не из списка 10 самых часто встречающихся киностудий, но получавшая господдержку, то она будет заменена два значения "другие" и "гп-другие"

```python
# Массив для значений, которые не вышло преобразовать
film_studio_highlight_top_exceptions = []


# Функция, которая оставляет только 10 самых часто встречающихся киностудий. Если часто-встречающаяся киностудия получала господдержку, то она будет иметь префикс "гп-". Если киностудия не входит в 10 самых часто встречающихся, но получала господдержку, то она будет заменена два значения "другие" и "гп-другие". Если киностудия не входит в 10 самых часто встречающихся и не получала господдержку, то она будет заменена на "другие".
def highlight_top_studios(s):
    try:
        processed_elements = []
        if isinstance(s, list):
            for element in s:
                if element in top_10_studios and element in gos_sup_studios_set:
                    processed_elements.append("гп-" + element)
                elif element in top_10_studios:
                    processed_elements.append(element)
                elif element in gos_sup_studios_set:
                    processed_elements.append("другие")
                    processed_elements.append("гп-другие")
                else:
                    processed_elements.append("другие")
            return processed_elements
        if isinstance(s, str):
            if s in top_10_studios and s in gos_sup_studios_set:
                return ["гп-" + s]
            elif s in top_10_studios:
                return [s]
            elif s in gos_sup_studios_set:
                return ["другие", "гп-другие"]
            else:
                return ["другие"]
        return pd.NA
    except Exception as e:
        # Вывод 10 первых ошибок
        print(f"Ошибка в строке {s}, {e}")
        film_studio_highlight_top_exceptions.append(s)
        raise Exception(f"Ошибка в строке {s}") from e


# Функция для запуска кода highlight_top_studios через функцию разрешающую запустить код только один раз за запуск тетради
def code_to_run():
    # Переменная которую надо изменить
    global data
    # Выполняемая функция
    data["film_studio"] = data["film_studio"].apply(
        lambda x: safe_process_value(
            x, film_studio_highlight_top_exceptions, highlight_top_studios
        )
    )


# Запуск кода code_to_run через функцию, разрешающую запустить код только один раз за запуск тетради. Сохранение результата в переменную, позволяющую сделать запись о том, что ячейка была запущена
highlight_top_studios_used = run_code_only_once(
    code_to_run, "highlight_top_studios_used"
)

# Вывод первых 10 необработанных значений
print(
    f"Сообщение. Первые 10 необработанных значений: {film_studio_highlight_top_exceptions[:10]}, всего не удалось обработать: {len(film_studio_highlight_top_exceptions)}. \n"
)

# Вывод первых 10 значений столбца `film_studio`
display(data["film_studio"].head(10))
a = 5

b = 3
# Добавления записи в случае, если ячейка была запущена. Если бы в ходе выполнения кода ячейки были ошибки, последняя строка не была бы запущена
if highlight_top_studios_used:
    cell_run_flags.add("highlight_top_studios_used")
```

```python
# Массив для значений, которые не вышло преобразовать
director_compress_name_exceptions = []


# Функция, которая заменяет имена режисеров на первую букву имени и точку. Работает для строк и для массивов строк. Если тип данных иной, то возвращает pd.NA
def compress_name(s):
    try:
        # Обработка если s строка
        if isinstance(s, str):
            if not ("." in s) and (" " in s):
                s = s.split(" ")
                s = s[0][:1] + "." + s[1]
                return s
            return s
        # Обработка если s список
        if isinstance(s, list):
            processed_elements = []
            for element in s:
                if not ("." in element) and (" " in element):
                    element = element.split(" ")
                    element = element[0][:1] + "." + element[1]
                    processed_elements.append(element)
                else:
                    processed_elements.append(element)
            return processed_elements
        # Обработка если s не строка и не список
        return pd.NA
    except Exception as e:
        print(f"Ошибка при обработке {s} {e}")
        raise Exception(f"Ошибка при обработке {s}")


# Применение функции
data["director"] = data["director"].apply(
    lambda x: safe_process_value(x, director_compress_name_exceptions, compress_name)
)

# Вывод первых 10 необработанных значений
print(
    f"Сообщение. Первые 10 необработанных значений: {director_compress_name_exceptions[:10]}, всего не удалось обработать: {len(director_compress_name_exceptions)}.\n"
)

# Вывод первых 5 значений столбца `director`
data["director"].head(5)
```

Добавим в сводную таблицу со сборами `year_grouped_box_office` столбцы, в котором общий сборы будут округлены до 10 миллионов, если они больше 100 млн и до 100 тысяч, если они больше 1 млн. Также представим миллионы и миллиарды в виде десятичных дробей

```python
# Функция для округления суммы сборов и записи в новом виде
def round_box_office(x):
    if x > 100000000:
        return f"{round(x / 100000000, 1)} млрд"
    elif x > 1000000:
        return f"{round(x / 1000000, 1)} млн"
    elif x > 1000:
        return f"{round(x / 1000, 1)} тыс"
    else:
        return x


# Применение функции к столбцу `total_box_office`
year_grouped_box_office["total_box_office_rounded"] = year_grouped_box_office[
    "total_box_office"
].apply(round_box_office)

# Приеменение функции к столбцу `mean_film_box_office`
year_grouped_box_office["mean_film_box_office_rounded"] = year_grouped_box_office[
    "mean_film_box_office"
].apply(round_box_office)

# Приеменение функции к столбцу `median_film_box_office`
year_grouped_box_office["median_film_box_office_rounded"] = year_grouped_box_office[
    "median_film_box_office"
].apply(round_box_office)

# Вывод первых 5 строк
year_grouped_box_office
```

Получим сводную таблицу с данными о количестве фильмов за год и количестве фильмов, получивших государственную поддержку, по годам

```python
# Получение сводной таблицы
year_grouped_gos_sup_number = data.pivot_table(
    index="show_start_year", values="puNumber", aggfunc=["count"]
)

# Переименование столбца
year_grouped_gos_sup_number.columns = ["film_number"]

# Считаем количесвтом фильмов, не получивших государственную поддержку
year_grouped_gos_sup_number["gos_sup_number"] = data.pivot_table(
    index="show_start_year",
    values="financing_source_отсутствует",
    aggfunc="sum",
    fill_value=0,
)

# Считаем количество фильмов получивших государственную поддержку
year_grouped_gos_sup_number["gos_sup_number"] = (
    year_grouped_gos_sup_number["film_number"]
    - year_grouped_gos_sup_number["gos_sup_number"]
)

# Добавление столбца с годом как отдельного столбца
year_grouped_gos_sup_number["show_start_year_col"] = year_grouped_gos_sup_number.index

# Посчитаем процент фильмов, получивших государственную поддержку
year_grouped_gos_sup_number["gos_sup_percent"] = (
    year_grouped_gos_sup_number["gos_sup_number"]
    / year_grouped_gos_sup_number["film_number"]
    * 100
).round(1)

# Вывод первых 5 строк
year_grouped_gos_sup_number.head(5)
```

Посчитаем сколько фильмов каждого типа получили государственную поддержку и сколько фильмов каждого типа было всего

```python
# Получение сводной таблицы
type_grouped_gos_sup_number = data.pivot_table(
    index="type", values="puNumber", aggfunc=["count"]
)

# Добавление колонки фильмов этого типа получивших государственную поддержку
type_grouped_gos_sup_number["gos_sup_number"] = (
    data.pivot_table(
        index=["type", "financing_source_отсутствует"],
        values="puNumber",
        aggfunc="count",
    )
    .query("financing_source_отсутствует == 0")
    .reset_index()
    .drop(columns="financing_source_отсутствует")
    .set_index("type")
)

# Переименование столбцов
type_grouped_gos_sup_number.columns = ["film_number", "gos_sup_number"]

# Добавление столбца с типом фильма как отдельного столбца
type_grouped_gos_sup_number["type_col"] = type_grouped_gos_sup_number.index

# Посчитаем процент фильмов каждого типа, получивших государственную поддержку
type_grouped_gos_sup_number["gos_sup_percent"] = (
    type_grouped_gos_sup_number["gos_sup_number"]
    / type_grouped_gos_sup_number["film_number"]
    * 100
).round(1)

# Заполнение пропусков нулями так как если в ячейке NaN, то это значит поддержки фильмов такого типа не было
type_grouped_gos_sup_number = type_grouped_gos_sup_number.fillna(0)

# Вывод первых 10 строк
type_grouped_gos_sup_number
```

```python
# Список столбцов часто встречающихся жанров
genre_columns = [
    "genre_биография",
    "genre_боевик",
    "genre_детектив",
    "genre_драма",
    "genre_другие",
    "genre_комедия",
    "genre_криминал",
    "genre_мелодрама",
    "genre_мультфильм",
    "genre_неизвестно",
    "genre_приключения",
    "genre_семейный",
    "genre_триллер",
    "genre_ужасы",
    "genre_фантастика",
    "genre_фэнтези",
]

# Количество фильмов каждого жанра
genre_grouped_gos_sup_number = pd.DataFrame(
    data.loc[:, genre_columns].sum(), columns=["total_film_number"]
)

# Количество фильмов каждого жанра, получивших государственную поддержку
genre_grouped_gos_sup_number["gos_sup_number"] = (
    data.query("financing_source_отсутствует == 0").loc[:, genre_columns].sum()
)

# Процент фильмов каждого жанра, получивших государственную поддержку
genre_grouped_gos_sup_number["gos_sup_percent"] = (
    genre_grouped_gos_sup_number["gos_sup_number"]
    / genre_grouped_gos_sup_number["total_film_number"]
    * 100
).round(1)

# Вывод таблицы
genre_grouped_gos_sup_number
```

Создадим сводную таблицу с данными о количестве фильмов каждой страны и количестве фильмов каждой страны, получивших государственную поддержку. Найдем процент фильмов каждой страны, получивших государственную поддержку

```python
# Список столбцов часто встречающихся стран
country_columns = [
    "country_бельгия",
    "country_великобритания",
    "country_германия",
    "country_другие",
    "country_испания",
    "country_италия",
    "country_канада",
    "country_россия",
    "country_ссср",
    "country_сша",
    "country_франция",
]

# Получение сводной таблицы с общим количеством фильмов каждой страны
country_grouped_gos_sup_number = pd.DataFrame(
    data.loc[:, country_columns].sum(), columns=["total_film_number"]
)

# Добавление столбца с количеством фильмов каждой страны, получивших государственную поддержку
country_grouped_gos_sup_number["gos_sup_number"] = (
    data.query("financing_source_отсутствует == 0").loc[:, country_columns].sum()
)

# Добавление столбца с процентом фильмов каждой страны, получивших государственную поддержку
country_grouped_gos_sup_number["gos_sup_percent"] = (
    country_grouped_gos_sup_number["gos_sup_number"]
    / country_grouped_gos_sup_number["total_film_number"]
    * 100
).round(1)

# Вывод таблицы
country_grouped_gos_sup_number
```

Знаем, что всего государственную поддержку получило 250 фильмов. То что в сумме поддержанных фильмов каждой страны больше 250 говорит о том, что один фильм может быть снят в нескольких странах. Например, фильм "Джентльмены" может быть снят в Великобритании и США.

Построим столбчатый график процентов фильмов каждой страны, получивших государственную поддержку

```python
# Создание фигуры
fig = go.Figure()

# Выберем данные со странами, которые хотя бы раз получили государственную поддержку
country_grouped_gos_sup_number_not_null = country_grouped_gos_sup_number.query(
    "gos_sup_number > 0"
)

# Добавление столбчатых графиков
fig.add_trace(
    go.Bar(
        x=country_grouped_gos_sup_number_not_null.index,
        y=country_grouped_gos_sup_number_not_null["gos_sup_percent"],
        name="Процент фильмов, получивших государственную поддержку",
        text=(
            [
                f"{i}%"
                for i in country_grouped_gos_sup_number_not_null["gos_sup_percent"]
            ]
        ),
        textposition="auto",
        textfont_size=15,
    )
)

# Оформление графика
fig.update_layout(
    title="Процент фильмов каждой страны, получивших государственную поддержку",
    xaxis_title="Страна",
    yaxis_title="Процент",
    title_x=0.5,  # Позиционоирование заголовка
)

# Вывод графика
fig.show()
```

Меньше всего господдержки получили фильмы из Канады, Испании, США, СССР, Великобритании (менее 1% или вовсе 0%), а больше всего фильмы из России (14.1%).

Посмотрим на взаимосвязь источника финансирования и жанра фильма

```python
# Список столбцов часто встречающихся жанров
genre_columns = [
    "genre_биография",
    "genre_боевик",
    "genre_детектив",
    "genre_драма",
    "genre_другие",
    "genre_комедия",
    "genre_криминал",
    "genre_мелодрама",
    "genre_мультфильм",
    "genre_неизвестно",
    "genre_приключения",
    "genre_семейный",
    "genre_триллер",
    "genre_ужасы",
    "genre_фантастика",
    "genre_фэнтези",
]

# Количество фильмов каждого жанра, получивших государственную поддержку
genre_financing_source_pivot_table = pd.DataFrame(
    data.query("financing_source_отсутствует == 0").loc[:, genre_columns].sum(),
    columns=["film_number"],
)


# Добавление колонок с источником финансирования и количеством фильмов во всех жанрах
genre_financing_source_pivot_table["source_фонд_кино"] = (
    data.query("financing_source_фонд_кино  == 1").loc[:, genre_columns].sum()
)
genre_financing_source_pivot_table["source_министерство_культуры"] = (
    data.query("financing_source_министерство_культуры == 1")
    .loc[:, genre_columns]
    .sum()
)
genre_financing_source_pivot_table["source_министерство_культуры_фонд_кино"] = (
    data.query(
        "financing_source_министерство_культуры == 1 & financing_source_фонд_кино == 1"
    )
    .loc[:, genre_columns]
    .sum()
)

genre_financing_source_pivot_table.head(15)
```
